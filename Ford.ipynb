### Variable Description


Variables Description
duration_sec (Numerical):

start_time (Datetime):

Description: The timestamp indicating when the trip started.

Format: YYYY-MM-DD HH:MM:SS.

Use: This column is helpful for time-based analysis, such as identifying peak trip times or trends based on the day of the week, hour of the day, etc.

end_time (Datetime):

Description: The timestamp indicating when the trip ended.

Format: YYYY-MM-DD HH:MM:SS.

Use: This, along with start_time, is essential for calculating the trip duration and analyzing the trip behavior throughout the day.

start_station_id (Categorical):

Description: Unique identifier for the station where the trip started.

Use: This column helps in identifying the specific station and is useful for station-level analysis, such as tracking station usage and frequency of bike pickups.

start_station_name (Categorical):

Description: The name of the station where the trip started.

Use: Helps identify where trips are initiated, useful for spatial analysis of station popularity and the flow of trips.

start_station_latitude (Numerical):

Description: The latitude coordinate of the starting station.

Use: This is part of the spatial data that can be used for mapping trips and analyzing geographic distribution.

start_station_longitude (Numerical):

Description: The longitude coordinate of the starting station.

Use: Works with start_station_latitude to determine the exact location of the station for geographic analysis.

end_station_id (Categorical):

Description: Unique identifier for the station where the trip ended.

Use: Like start_station_id, this helps in identifying the destination station for each trip.

end_station_name (Categorical):

Description: The name of the station where the trip ended.

Use: This column helps analyze the destinations of trips, which can be important in understanding flow patterns between stations.

end_station_latitude (Numerical):

Description: The latitude coordinate of the ending station.

Use: This can be used alongside end_station_longitude to visualize or map out the routes taken by the bikes.

end_station_longitude (Numerical):

Description: The longitude coordinate of the ending station.

Use: Helps in plotting the ending locations for trip visualizations.

bike_id (Categorical):

Description: The unique identifier of the bike used during the trip.

Use: This is important for tracking bike usage, maintenance schedules, and ensuring there are no issues with individual bikes.

user_type (Categorical):

Description: Indicates whether the user is a "Subscriber" or a "Customer".

Values: 'Subscriber', 'Customer'.

Use: This helps to segment users and analyze how different types of users interact with the service, e.g., frequency of usage, trip duration, etc.

member_birth_year (Numerical):

Description: The birth year of the rider.

Use: This can be used to derive the age of the user (by subtracting the birth year from the current year) and analyze trends based on age groups.

member_gender (Categorical):

Description: The gender of the rider.

Values: 'Male', 'Female', 'Other' (with possible missing values).

Use: This is helpful in identifying any gender-based usage patterns, such as trip frequency or duration.

bike_share_for_all_trip (Categorical):

Description: Indicates whether the trip was eligible for the "Bike Share for All" program (a discounted program for low-income members).

Values: 'True', 'False'.

#**3. Data Wrangling**

```python
# Write your code to make your dataset analysis ready.

#Converting Date Time Columns into Date time format
ford["start_time"]=pd.to_datetime(ford['start_time'])
ford['end_time']=pd.to_datetime(ford['end_time'])

#Creating a New column based on existing data
ford['trip_duration_min']=ford['duration_sec']/60

#Converting user_type and member_gender to category type for memory optimization
ford['user_type']=ford['user_type'].astype('category')
ford['member_gender']=ford['member_gender'].astype("category")

print("\nData after wrangling:")
print(ford.head())
```

### What all manipulations have you done and insights you found?

### **Data Manipulations:**

1. **Datetime Conversion:**
   - Converted **start_time** and **end_time** columns to datetime format for accurate time-based analysis (e.g., calculating trip durations, identifying patterns over time).

2. **Trip Duration Calculation:**
   - Created a new column **trip_duration_min** by converting **duration_sec** from seconds to minutes for easier analysis and comparison of trip lengths.

3. **Memory Optimization:**
   - Converted **user_type** and **member_gender** columns to categorical data type, reducing memory usage and improving processing efficiency.

4. **Missing Data Handling:**
   - Identified missing values in **member_birth_year** (7,839 missing) and **member_gender** (7,801 missing). These will need imputation or removal before further analysis.

5. **Duplicate Check:**
   - Checked for duplicate records, confirming that there were no duplicates in the dataset.

### **Insights Found:**

1. **Trip Duration Variability:**
   - The **trip_duration_sec** column shows significant variation, suggesting that trip lengths range from very short to long durations. This can be categorized into short, medium, and long trips for more detailed analysis.

2. **User Types:**
   - The dataset includes both **Subscribers** and **Customers**, with Subscribers likely being frequent users. This suggests differing usage patterns between the two groups (e.g., frequency and duration of trips).

3. **Station Analysis:**
   - **Start and End Station IDs/Names** reveal the popularity and activity at different stations. This can be used to identify high-traffic stations and suggest improvements in bike distribution.

#### Chart - 8 - Bar Chart

```python
# Chart - 8 visualization code
# Calculate rider age
ford['rider_age'] = 2018 - ford['member_birth_year']  # Dataset year is 2018

# Create age groups
bins = [0, 18, 25, 35, 45, 55, 65, 100]
labels = ['<18', '18-24', '25-34', '35-44', '45-54', '55-64', '65+']
ford['age_group'] = pd.cut(ford['rider_age'], bins=bins, labels=labels, right=False)

# Group by age group
age_group_counts = ford['age_group'].value_counts().reindex(labels)

# Plotting
plt.figure(figsize=(10,6))
sns.barplot(x=age_group_counts.index, y=age_group_counts.values, palette='coolwarm')

plt.title('Number of Trips by Rider Age Group', fontsize=16)
plt.xlabel('Age Group', fontsize=14)
plt.ylabel('Number of Trips', fontsize=14)
plt.grid(axis='y')
plt.show()
```
